```python
#Importar biblioteca pandas

import pandas as pd
```

```python
combustiveis_df = pd.read_excel('ca-2021-02.xlsx')
display(combustiveis_df.head())
```

```python
#Inserção simples de dados em uma nova coluna
combustiveis_df['Ativo'] = True
display(combustiveis_df.head())
```

```python
# E se eu quiser uma coluna nova chamada "OBS" que preencha automaticamente "Melhor cidade" quando o municipio for Joinville SC
combustiveis_df['Obs'] = ["Melhor Cidade" if municipio == 'JOINVILLE' else '' for municipio in combustiveis_df['Municipio']]
display(combustiveis_df.loc[combustiveis_df['Municipio'].isin(['JOINVILLE','ARAPIRACA', 'CAMPINAS', 'SALTO']), ['Municipio', 'Obs']])
```

```python
#Exemplo da sala
import numpy as np

combustiveis_df['Valor de Venda - Status'] = np.where(combustiveis_df['Valor de Venda'] > 6.5, 'Caro', 'Barato')
display(combustiveis_df)
```

```python
#Calcular número de postos de combustíveis por habitantes temos na amostragem de combustíveis nov/2021 (Já incluso minha cidade)
num_habitantes_df = pd.read_csv('ibge_num_habitantes_estimado.csv')
num_habitantes_df.rename(columns={"Estado":"Estado - Sigla"}, inplace=True)
display(num_habitantes_df)
```

```python
#Média de habitantes pela planilha num_habitantes_df
media_habitantes_df = num_habitantes_df[['NumHabitantes2021']].groupby(by=['NumHabitantes2021']).mean()
display(media_habitantes_df)
```

```python
#Faz um MERGE dos dois dataframes
colunas = ['Municipio', 'Estado - Sigla']
merge_df = combustiveis_df.merge(num_habitantes_df, how="inner", on=colunas)
display(merge_df)
print(merge_df.info())
```

```python
#Destruir coluna completamente vazia
merge_df.dropna(axis='columns', inplace=True)
print(merge_df.info())
```

```python
colunas=['Regiao - Sigla', 'Nome da Rua', 'Numero Rua', 
         'Bairro', 'Cep', 'Produto', 'Data da Coleta', 'Valor de Venda',
         'Unidade de Medida', 'Bandeira', 'Ativo', 'Valor de Venda - Status', 'Obs', 'Complemento', 'Valor de Compra']
merge_df.drop(labels=colunas, axis=1, inplace=True)
print(merge_df.info())
```

```python
display(merge_df.head(100))
```

```python
#Remover as linhas duplicadas
merge_df.drop_duplicates(inplace=True)
display(merge_df.head(100))
```

```python
#Agrupar e contar quantos postos tem na cidade
postos_por_municipio_df = merge_df.groupby(by=['Estado - Sigla', 'Municipio', 'NumHabitantes2021']).count()
postos_por_municipio_df.drop('CNPJ da Revenda', axis=1, inplace=True)
postos_por_municipio_df.rename(columns={"Revenda": "Numero de Postos"}, inplace=True)
display(postos_por_municipio_df)

```
